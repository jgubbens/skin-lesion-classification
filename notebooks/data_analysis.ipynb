Class Distributions

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

data_dir = "../data/raw"
classes = sorted([c for c in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, c))])

# Count images per class
class_counts = {cls: len(os.listdir(os.path.join(data_dir, cls))) for cls in classes}

print(class_counts)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(class_counts.keys(), class_counts.values(), color='blue')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.tight_layout()
plt.show()
```
